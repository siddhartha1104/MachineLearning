```python
from selenium import webdriver
import chromedriver_binary
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
joblinks = pd.read_csv('../datasets/link_by_areas.csv')
joblinks
```

```python
urls = joblinks['link'].tolist()
urls
```

```python
from yarl import URL
```

```python
n = []
for i in urls:
    n.append(URL(i).path)
```

```python
n
```

```python
m = []

for i in urls:
    m.append(URL(i).query_string)
```

```python
m
```

```python
gen_urls = []

for i in range(len(n)):
    url = 'https://www.naukri.com/'+n[i]+'-{}?'+m[i]
    gen_urls.append(url)
```
