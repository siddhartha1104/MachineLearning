```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../datasets/auto-mpg.csv")
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.head(2)
```

```python
df['origin'] = df['origin'].replace({1:'america', 2:'europe', 3:'asia'})
```

```python
df.head(2)
```

```python
df = df.drop('car name',axis = 1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
hpisDigit = pd.DataFrame(df.horsepower.str.isdigit())
```

```python
df[hpisDigit['horsepower'] == False]
```

```python
df = df.replace('?', np.nan)
df[hpisDigit['horsepower'] == False]
```

```python
df.isna().sum()
```

```python
medianCheack = df.drop('origin', axis = 1)
medianCheack
```

```python
medianCheack.median()
```

# to replace NaN value with -> median

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df.isna().sum()
```

```python
df[hpisDigit['horsepower'] == False]
```

# above values are replaced with median value

```python
df.dtypes
```

# we have to conver Horsepower -> object to --> float

```python
df['horsepower'] = df['horsepower'].astype('float64')
```

```python
df.dtypes
```

# to create dummies variables

```python
df = pd.get_dummies(df,['origin'])
```

```python
df
```

```python
df.dtypes
```

```python
df
```

```python
x = df.drop('mpg', axis = 1)
y = df['mpg']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
y_test
```

```python
model = LinearRegression()
model
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train,y_train)*100
```

```python
model.score(x_test, y_test)*100
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict(x)
```
