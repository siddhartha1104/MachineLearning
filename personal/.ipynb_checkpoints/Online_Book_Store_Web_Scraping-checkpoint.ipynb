```python
import pandas as pd
import requests
import bs4
```

# Creating empty lists

```python
# Creating empty lists
pages = []
prices = []
ratings = []
title = []
urls_links = []

no_of_pages = 1 #number of pages to be selected

for i in range (1, no_of_pages+1):
    url = ('https://books.toscrape.com/catalogue/page-{}.html'.format(i))

    pages.append(url)
```

```python
print("No. of pages: ", len(pages))
print(pages)
```

```python
for i in pages:
    webData = requests.get(i)
    bs4_webData = bs4.BeautifulSoup(webData.text, 'html.parser')

    print(bs4_webData)
```

```python
print(bs4_webData.prettify())
```

```python
pages
```

```python
print(bs4_webData)
```

# Fingind Titles(book names)

```python
# Print all titles

for i in bs4_webData.findAll('h3'):
    t = i.getText()
    title.append(t)
```

```python
title
```

# Pinding Price

```python
for i in bs4_webData.find_all('p', class_ = 'price_color'):
    price = i.getText()
    prices.append(price)
    
    
    
```

```python
prices
```

# Finding Ratings

```python
bs4_webData.find_all('p', class_ = 'star-rating')
```

```python
# getting the ratings
# k = class
# v = ['star-rating', 'one']

for i  in bs4_webData.find_all('p', class_ = 'star-rating'):
    for k,v in i.attrs.items():
        star = v [1]
        ratings.append(star)
        print(v)
```

```python
ratings
```

# Finding URls

```python
divs = bs4_webData.find_all('div', class_ = 'image_container')
# print(divs)

for url in divs:
    tagss = url.find('img', class_ = 'thumbnail')
    # print(tagss)
    links = 'https://books.toscrape.com/' + str(tagss['src'])
    # print(links)
    newlinks = links.replace('..','') # to get rid of the three-dots
    # print(newlinks)
    urls_links.append(newlinks)
```

```python
urls_links
```
